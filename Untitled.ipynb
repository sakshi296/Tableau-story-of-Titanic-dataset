## Summary

This project has different types of graphs. All the graphs shows different findings. It shows, How did the number as well as pecentage of survivors affected by age-groups? Which gender has more survival rate? and many more. All the findings in the graphic representation involves around the number of survivors. 

## Design

I have used titanic datset for visualization. I have worked on Tableau and made interesting visualizations. While exploring data , I found that females had more survival rate than males. The second thing that I noticed was, people with age-group 16 to 30 had more survival rate than other age groups. According to the dataset I had, I decided to use bar charts to visualize those assumptions which would be displayed further down below. 

### Data visualization

I considered using multiple charts(bar charts, pie-charts, stacked-charts etc.) to test if this is a good way to visualize and stress on important point. Here is the description of my all the designs.

* **Survival By gender** 

Chart type -> Bar Chart 

**Reason for choosing bar chart**

I have used bar chart for this visualization because these charts are useful for displaying data that is classified into nominal or odinal categories. If you have comparative data that you would like to represent through a chart then a bar chart would be the best option. Therefore I have used bar charts for this.

Here I have used the total records and found the number of survivors according to gender. According to the graph, I have noticed that number of females survived were more as compared to males. As seen in the Titanic movie that females were given life-jackets first. Moreover this can be one reason that more number of females were saved.

* **Survival by Age - group**
 
Chart type -> Stacked bar chart

**Reason for choosing stacked bar chart**

I have used stacked bar chart for this visualization because they are great for showing the total because they visually aggregate all of the categories in a group. Stacking also indicates a part to whole relationship.

Here I have divided ages in groups. I have made 5 age groups of sizes 0-15, 16-30, 31-45, 46-60 and 61+. Then I found the survival rate for all the people falling in the these age-groups. I noticed that 16-30 age grouped people had higher survival rate as compared to others. 

Extending the above graph, I have also found how did the age-group and gender affected the survival rate. 

* **Survival rate for different embarkments**

Chart type -> Pie chart

**Reason for choosing pie chart**

I have used pie chart for this visualization because the data here (embarkments) is categorical therefore if we have categorical data then using a pie chart would work really well as each slice can represent a different category. Therefore I have used pie chart here.

As it can be seen in the dataset, we have 3 embarkments site Southhampton,UK(S), Cherbourg,FR(C) and Queenstown,IRE(Q). According to this I found the survival rate for all these 3 embarkments. This can be seen more clearly in the visualizatin

## Final Story

After taking the feedback I have made some changes. As asked in the feedback that relatives survival rate was missing. Therefore I have made another graph of survival rate for relatives. As asked in the feedback I tried to add everything such that it can become easier for the viewer to understand. 

For making this graph, first of all I found that how many people had relatives i.e members present in a family. As given in the dataset we have measures for Parent-Children and Sibling-Spouse. From this information I made another field named "Have relatives" where I calculated total of Parent-Children and Sibling-Spouse. And after doing the calculation I plotted the graph between percentage of total(Sum of records) and relatives. I have used bar charts for this because it is easy for making comparisions. 

The resulting graph indicates 2 coulums, one that have relatives and other that do not have relatives. And finally we can see the percentage of people survived in each column.
